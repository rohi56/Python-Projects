**random()** : Returns a random float in the range [0.0, 1.0)

**uniform(a, b)**	Returns a random float in the range [a, b]

**randint(a, b)**	Returns a random integer in the range [a, b]

**randrange(start, stop[, step])**	Returns a random integer from range(start, stop, step)

**gauss(mu, sigma)**	Returns a float from a normal distribution with mean mu and standard deviation sigma

**normalvariate(mu, sigma)**	Similar to gauss(), but different underlying algorithm

**betavariate(alpha, beta)**	Beta distribution

**expovariate(lambd)**	Exponential distribution

**gammavariate(alpha, beta)**	Gamma distribution

**lognormvariate(mu, sigma)**	Log-normal distribution

**triangular(low, high, mode)**	Triangular distribution

**vonmisesvariate(mu, kappa)**	Von Mises distribution

**weibullvariate(alpha, beta)**	Weibull distribution

## Explore math module

math module is another interesting and useful module to help you deal with math. Explore and learn by yourself.

**Basic Math Functions**

| Function | Description |
|----------|-------------|
| `math.ceil(x)` | Returns the smallest integer ≥ `x` |
| `math.floor(x)` | Returns the largest integer ≤ `x` |
| `math.trunc(x)` | Truncates (removes) the decimal part of `x` |
| `math.fabs(x)` | Absolute value of `x` (float) |
| `math.factorial(x)` | Returns the factorial of `x` |
| `math.fmod(x, y)` | Returns remainder of `x / y` (like `x % y` but supports floats) |
| `math.remainder(x, y)` | IEEE 754-style remainder of `x / y` |
| `math.gcd(x, y)` | Greatest common divisor of `x` and `y` |
| `math.lcm(x, y)` | Least common multiple (Python 3.9+) |


**Exponents and Logarithms**

| Function | Description |
|----------|-------------|
| `math.exp(x)` | `e` raised to the power `x` |
| `math.log(x[, base])` | Natural log of `x`, or log base `base` |
| `math.log10(x)` | Base-10 logarithm |
| `math.log2(x)` | Base-2 logarithm |
| `math.pow(x, y)` | `x` raised to the power `y` (returns float) |
| `math.sqrt(x)` | Square root of `x` |


**Trigonometric Functions**

| Function | Description |
|----------|-------------|
| `math.sin(x)` | Sine of `x` (in radians) |
| `math.cos(x)` | Cosine of `x` |
| `math.tan(x)` | Tangent of `x` |
| `math.asin(x)` | Arc sine of `x` (result in radians) |
| `math.acos(x)` | Arc cosine of `x` |
| `math.atan(x)` | Arc tangent of `x` |
| `math.atan2(y, x)` | Arc tangent of `y/x`, considering the quadrant |
| `math.radians(x)` | Converts degrees to radians |
| `math.degrees(x)` | Converts radians to degrees |


** Hyperbolic Functions**

| Function | Description |

| `math.asinh(x)` | Inverse hyperbolic sine |
