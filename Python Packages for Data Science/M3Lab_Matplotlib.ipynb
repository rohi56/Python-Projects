## Set up the environment

### Task: import packages and rename them accordingly

````python
# your code is here
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


sns.set(style="whitegrid")
````

## Load the dataset

````python
# your code is here

data = sns.load_dataset('tips')
print(data.head())
````

## Play with distribution plots
You can show the data distribution for some attributes interest you.

````python
# your code is here
plt.figure(figsize=(8, 6))
sns.histplot(data['total_bill'], bins=20, kde=True)
plt.title('Distribution of Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.show()
````

## Play with relational plots
You can show how two or more attributes are associated.

````python
# your code is here
plt.figure(figsize=(8, 6))
sns.scatterplot(data=data, x='total_bill', y='tip', hue='sex')
plt.title('Tip vs Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
````

## Play with other plots
You can play with other plots we learned here. Be as creative as you can and see how the visualization helps you to understand the data better.

````python
# your code is here
plt.figure(figsize=(8, 6))
sns.boxplot(data=data, x='day', y='tip', hue='sex')
plt.title('Tips by Day and Gender')
plt.xlabel('Day')
plt.ylabel('Tip')
plt.show()
````

## Additional Challenge
You may want to create new columns based on existing data using Pandas. You can play with the new columns as well.

````python
# your code is here
data['tip_percent'] = (data['tip'] / data['total_bill']) * 100

# Visualize it
plt.figure(figsize=(8, 6))
sns.histplot(data['tip_percent'], bins=20, kde=True, color='purple')
plt.title('Distribution of Tip Percentage')
plt.xlabel('Tip Percentage')
plt.ylabel('Frequency')
plt.show()
````
